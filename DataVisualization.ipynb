```python
x=[1,2,3]
y=[1,4,9]
z=[10,5,0]
plt.plot(x,y)
plt.plot(y,z)
plt.title('Test Plot')
plt.xlabel('x')
plt.ylabel('y and z')
plt.legend(["This is x","This is y"])
plt.show()
```

```python
df1 = pd.read_csv('/home/valerie/Everything Data/Personal/countries.csv')
```

```python
df1
```

```python
us = df1[df1.country == 'United States']
```

```python
us
```

```python
china = df1[df1.country == 'China']
```

```python
china
```

```python
plt.plot(us.year,us.population / 10**6)
plt.plot(china.year,china.population / 10**6)
plt.legend(["US","China"])
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```
